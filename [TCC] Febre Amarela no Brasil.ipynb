A Febre Amarela é uma doença infecciosa transmitida por mosquitos, que afeta tanto humanos quanto primatas não-humanos, ainda, ela pode ser propagada por outras razões como a falta de instalações sanitárias adequadas e condições insalubres. A análise dos dados de vacinação é relevante para identificar regiões com baixa cobertura vacinal e implementar medidas de prevenção e controle da doença.

A utilização de técnicas de Machine Learning pode ser útil para prever se os pacientes infectados com a doença chegarão a óbito ou não, fornecendo informações relevantes para o controle da doença.

Ao final deste projeto de TCC, espera-se obter um modelo de Machine Learning preciso para classificar os casos de Febre Amarela e prever a mortalidade em pacientes infectados, utilizando tanto os dados epidemiológicos e de epizootias quanto as informações das instalações sanitárias, dados de vacinação, socioeconômicos e de população.

As bases de dados principais, com dados epidemiológicos e epizootias da doença, estão disponível neste [link](https://opendatasus.saude.gov.br/dataset/febre-amarela-em-humanos-e-primatas-nao-humanos). Essa fonte forneceu informações cruciais para análise da doença.

Ademais, outras quatro fontes de dados foram utilizadas para enriquecer o conjunto de dados. A primeira fonte refere-se às informações das instalações sanitárias do país, obtidas através deste [link](https://datasus.saude.gov.br/saneamento-censos-1991-2000-e-2010). A segunda fonte é composta por dados de vacinação contra a Febre Amarela, que disponibiliza informações sobre a cobertura vacinal por município, coletados a partir desse [link](http://tabnet.datasus.gov.br/cgi/dhdat.exe?bd_pni/cpnibr.def). A terceira fonte é referente aos dados populacionais do Brasil obtidos no [link](https://www.ibge.gov.br/estatisticas/sociais/populacao/9103-estimativas-de-populacao.html?edicao=17283&t=downloads) do IBGE. Por fim, a quarta fonte é o Atlas Brasil, uma plataforma que fornece informações socioeconômicas e ambientais de diferentes regiões do país, incluindo dados sobre saúde, educação, renda e meio ambiente. Essa fonte pode ser acessada através do [link](http://www.atlasbrasil.org.br) e forneceu dados adicionais para complementar a análise da Febre Amarela e seus possíveis fatores associados.

### 2.1 Tratamento e transformação dos dados

```python
# Deleta os valores nulos e converte a coluna para 'int64'
df_fa = df_fa.dropna(subset=['COD_MUN_LPI'])
df_fa['COD_MUN_LPI'] = df_fa['COD_MUN_LPI'].astype('int64')

# No dataset de macacos 
# Agrupa os dados pelo COD_MUN_OCOR e ANO_OCOR e conta o número de ocorrências de cada grupo
df_count = df_mc.groupby(['COD_MUN_OCOR', 'ANO_OCOR']).size().reset_index(name='MACACOS_NUM_CASOS')

# Cria o novo DataFrame com as colunas desejadas
df_mc = pd.DataFrame({
    'COD_MUN_LPI': df_count['COD_MUN_OCOR'],
    'ANO_OCOR': df_count['ANO_OCOR'],
    'MACACOS_NUM_CASOS': df_count['MACACOS_NUM_CASOS']
})
```

Está sendo realizado um tratamento nos dados dos dataframes df_fa e df_mc.

No dataframe df_fa, está sendo realizada a exclusão dos registros com valores nulos na coluna "COD_MUN_LPI" utilizando o método "dropna" e o parâmetro "subset" para especificar a coluna a ser verificada. Em seguida, a coluna "COD_MUN_LPI" está sendo convertida para o tipo 'int64' utilizando o método "astype". Essa etapa é importante para garantir que os dados estejam consistentes e prontos para a análise exploratória.

No dataframe df_mc, está sendo feito um agrupamento dos dados utilizando o método "groupby" para agrupar os dados pelo código do município de ocorrência (COD_MUN_OCOR) e ano de ocorrência (ANO_OCOR) e contar o número de ocorrências de cada grupo utilizando o método "size". Em seguida, está sendo criado um novo dataframe com as informações de interesse: o código do município de localização do primata infectado (COD_MUN_LPI), o ano de ocorrência (ANO_OCOR) e o número de casos (MACACOS_NUM_CASOS). Essa etapa é importante para simplificar e preparar os dados para a análise exploratória e para a construção de modelos de Machine Learning.

---

```python
# Renomeia a coluna 'Código Município' para 'COD_MUN_LPI'
df_is = df_is.rename(columns={'Código Município': 'COD_MUN_LPI'})
df_vc = df_vc.rename(columns={'Código Município': 'COD_MUN_LPI'})
df_pp = df_pp.rename(columns={'Código Município': 'COD_MUN_LPI'})
df_at = df_at.rename(columns={'Código Município': 'COD_MUN_LPI'})

# Junta as duas bases de dados pela coluna 'COD_MUN_LPI'
df = pd.merge(df_fa, df_is, on='COD_MUN_LPI', how='inner')
df = pd.merge(df, df_mc, on='COD_MUN_LPI', how='inner')
df = pd.merge(df, df_vc, on='COD_MUN_LPI', how='inner')
df = pd.merge(df, df_pp, on='COD_MUN_LPI', how='inner')
df = pd.merge(df, df_at, on='COD_MUN_LPI', how='inner')

# deleta os dados referente a datas anteriores a 2010
df.drop(df[df['ANO_IS'] < 2010].index, inplace=True)

# Renomear colunas
df = df.rename(columns={'Município': 'MUNICIPIO',
                        'Rede geral de esgoto ou pluvial': 'REDE_GERAL_ESGOTO_OU_PLUVIAL',
                        'Fossa séptica': 'FOSSA_SEPTICA',
                        'Fossa rudimendar': 'FOSSA_RUDIMENDAR',
                        'Vala': 'VALA',
                        'Rio, lago ou mar': 'RIO_LAGO_OU_MAR',
                        'Outro escoadouro': 'OUTRO_ESCOADOURO',
                        'Não tem instalação sanitária': 'NAO_TEM_INST_SANITARIA'})

# Altera valores "-" para NaN
cols_to_replace = ['REDE_GERAL_ESGOTO_OU_PLUVIAL', 
                   'FOSSA_SEPTICA', 'FOSSA_RUDIMENDAR', 
                   'VALA', 'RIO_LAGO_OU_MAR', 
                   'OUTRO_ESCOADOURO', 
                   'NAO_TEM_INST_SANITARIA']

df[cols_to_replace] = df[cols_to_replace].replace('-', np.nan)
```

O código está realizando diversas operações de manipulação e transformação de dados em pandas. Primeiro, ele renomeia a coluna 'Código Município' para 'COD_MUN_LPI' em várias bases de dados diferentes: df_is, df_vc, df_pp e df_at. Em seguida, ele junta todas essas bases de dados em um único dataframe df utilizando a coluna 'COD_MUN_LPI' como chave de junção.

Depois disso, o código deleta todas as linhas do dataframe que tenham valores de 'ANO_IS' menores que 2010. Em seguida, ele renomeia algumas colunas do dataframe, utilizando o método rename() do pandas.

Por fim, ele substitui todos os valores '-' por NaN em algumas colunas do dataframe utilizando o método replace() do pandas. As colunas que serão alteradas são 'REDE_GERAL_ESGOTO_OU_PLUVIAL', 'FOSSA_SEPTICA', 'FOSSA_RUDIMENDAR', 'VALA', 'RIO_LAGO_OU_MAR', 'OUTRO_ESCOADOURO' e 'NAO_TEM_INST_SANITARIA'.

Fazer essas etapas de limpeza e organização dos dados é importante para garantir que a análise e as conclusões feitas a partir dos dados sejam precisas e confiáveis.

---

```python
# Agrupa os dados de Taxa de Vacinação pela coluna UF_LPI
grouped = df.groupby(['UF_LPI'])[df.columns[25:37]]

# Calcula a média para cada grupo
means = grouped.mean()

# Substitui os valores faltantes pela média correspondente
df[df.columns[25:37]] = grouped.transform(lambda x: x.fillna(x.mean()))

# Trata valores NaN
tratar_valores_nan(df)

# Transformar a coluna 'IDADE' em uma variável numérica
df['IDADE'] = pd.to_numeric(df['IDADE'], errors='coerce')

# Converter as variáveis categóricas em variáveis numéricas
df['SEXO'] = df['SEXO'].map({'M': 0, 'F': 1}) # (M) Masculino (F) Feminino

# Transformar a coluna 'OBITO' em uma variável binária
df['OBITO'] = df['OBITO'].apply(lambda x: 1 if x == 'SIM' else 0)

# Converter as colunas categóricas para o tipo "category"
cat_cols = ['MACRORREG_LPI', 'UF_LPI', 'MUN_LPI', 'MONITORAMENTO_IS']
for col in cat_cols:
    df[col] = df[col].astype('category')

# Converter as colunas numéricas para o tipo "int64"
df.iloc[:, 25:37] = df.iloc[:, 25:37].astype('int64')
df.iloc[:, 15:22] = df.iloc[:, 15:22].astype('int64')
```

Estão sendo realizadas diversas transformações nos dados do conjunto de dados:

1. Agrupamento dos dados de Taxa de Vacinação pela coluna UF_LPI: os dados da coluna Taxa de Vacinação são agrupados pela coluna UF_LPI, que indica a unidade federativa do local de provável infecção.

2. Cálculo da média para cada grupo: a média dos valores de cada coluna do grupo é calculada. Essa é uma forma de preencher valores faltantes, que são substituídos pela média correspondente. Essa abordagem é importante para evitar a exclusão de dados que podem ser valiosos para a análise.

3. Tratamento de valores NaN: uma função de tratamento de valores NaN é aplicada aos dados do conjunto de dados. Essa é outra forma de preencher valores faltantes que podem ter passado despercebidos.

4. Transformação da coluna 'IDADE' em uma variável numérica: a coluna 'IDADE' é convertida em um tipo numérico para que possa ser utilizada em análises estatísticas.

5. Conversão das variáveis categóricas em variáveis numéricas: a variável 'SEXO' é convertida em uma variável numérica binária para que possa ser utilizada em análises estatísticas. Isso é importante porque muitas técnicas estatísticas só podem ser aplicadas a variáveis numéricas.

6. Transformação da coluna 'OBITO' em uma variável binária: a coluna 'OBITO' é convertida em uma variável binária para que possa ser utilizada em análises estatísticas. A conversão para uma variável binária é útil porque ela permite a aplicação de técnicas estatísticas específicas para dados binários.

7. Conversão das colunas categóricas para o tipo "category": as colunas categóricas são convertidas em uma categoria específica para que possam ser utilizadas em análises estatísticas. Essa conversão é importante porque ela reduz o espaço de memória utilizado pelos dados.

8. Conversão das colunas numéricas para o tipo "int64": as colunas numéricas são convertidas para o tipo "int64" para que possam ser utilizadas em análises estatísticas. Essa conversão é importante porque ela garante que os dados sejam do tipo numérico correto, o que pode ser importante para algumas técnicas estatísticas.

```python
# Quantidade de pessoas imunizadas no ano correspondente à coluna ANO_IS
df['IMUNIZACAO_NO_MUNICIPIO'] = df.apply(
    lambda row: row[str(int(row['ANO_IS']))], axis=1)

# Quantidade da população estimada no ano correspondente à coluna ANO_IS
coluna_fixa = "População_estimada_"
df['POPULACAO_NO_MUNICIPIO'] = df.apply(
    lambda row: row[coluna_fixa + str(int(row['ANO_IS']))], axis=1)

# Quantidade da cobertura vegetal natural no ano correspondente à coluna ANO_IS
coluna_fixa_cvn = "% de cobertura vegetal natural "
df['COBERTURA_VEGETAL_NATURAL'] = df.apply(
    lambda row: row[coluna_fixa_cvn + str(int(row['ANO_IS']))] 
    if 2013 <= row['ANO_IS'] <= 2017 else np.nan, axis=1)

# Quantidade da concentração dos focos de calor no ano correspondente à coluna ANO_IS
coluna_fixa_cfc = "Concentração dos focos de calor "
df['CONCENTRACAO_FOCOS_CALOR'] = df.apply(
    lambda row: row[coluna_fixa_cfc + str(int(row['ANO_IS']))] 
    if 2013 <= row['ANO_IS'] <= 2017 else np.nan, axis=1)
```

Nesta estágio, estão sendo criadas novas colunas no dataframe que irão armazenar informações relacionadas a diferentes variáveis do ambiente e da população dos municípios. Essas informações estão sendo extraídas de colunas específicas que estão identificadas por meio do ano de referência (coluna "ANO_IS").

A primeira coluna criada é a "IMUNIZACAO_NO_MUNICIPIO", que irá armazenar a quantidade de pessoas imunizadas no município no ano de referência. Essa informação é importante para avaliar a cobertura vacinal do município, que é um indicador fundamental para o controle de doenças e a prevenção de surtos.

A segunda coluna criada é a "POPULACAO_NO_MUNICIPIO", que irá armazenar a quantidade de pessoas estimadas que vivem no município no ano de referência. Essa informação é importante para calcular as taxas de incidência de doenças e avaliar a distribuição geográfica de determinados problemas de saúde.

As colunas "COBERTURA_VEGETAL_NATURAL" e "CONCENTRACAO_FOCOS_CALOR" irão armazenar informações relacionadas ao ambiente em que o município está inserido. Essas informações são importantes para avaliar o impacto ambiental sobre a saúde da população e identificar possíveis fatores de risco.

Em resumo, a criação dessas colunas é importante para que se possa realizar análises mais completas e aprofundadas sobre a relação entre saúde e ambiente nos municípios, e para que se possa identificar possíveis fatores de risco e desenvolver estratégias de prevenção e controle de doenças.

```python
# Calcula a taxa de vacinação para cada município em cada ano
df['TAXA_VACINACAO'] = df['IMUNIZACAO_NO_MUNICIPIO'] / df['POPULACAO_NO_MUNICIPIO']

# Converter as colunas e somar a faixa etária de Homens e Mulheres
df['H_0a19'] = df.loc[:, 'HOMEM0A4':'HOMEM15A19'].apply(sum, axis=1)
df['H_20a39'] = df.loc[:, 'HOMEM20A24':'HOMEM35A39'].apply(sum, axis=1)
df['H_40a59'] = df.loc[:, 'HOMEM40A44':'HOMEM55A59'].apply(sum, axis=1)
df['H_60ouMais'] = df.loc[:, 'HOMEM60A64':'HOMENS80'].apply(sum, axis=1)
df['M_0a19'] = df.loc[:, 'MULH0A4':'MULH15A19'].apply(sum, axis=1)
df['M_20a39'] = df.loc[:, 'MULH20A24':'MULH35A39'].apply(sum, axis=1)
df['M_40a59'] = df.loc[:, 'MULH40A44':'MULH55A59'].apply(sum, axis=1)
df['M_60ouMais'] = df.loc[:, 'MULH60A64':'MULHER80'].apply(sum, axis=1)

coluna_inicial = 'HOMEM0A4'
coluna_final = 'MULHER80'

todas_as_colunas = df.columns.tolist() 
indice_coluna_inicial = todas_as_colunas.index(coluna_inicial)  
indice_coluna_final = todas_as_colunas.index(coluna_final) 
colunas_para_excluir = todas_as_colunas[indice_coluna_inicial:indice_coluna_final+1]  

df.drop(colunas_para_excluir, axis=1, inplace=True)  # Excluir as colunas selecionadas

# Converter as colunas em tipo 'float'
df['SOBRE60'] = df['SOBRE60'].astype(str).str.replace(',', '.').astype(float)
df['GINI'] = df['GINI'].astype(str).str.replace(',', '.').astype(float)
df['IDHM'] = df['IDHM'].astype(str).str.replace(',', '.').astype(float)
df['IDHM_L'] = df['IDHM_L'].astype(str).str.replace(',', '.').astype(float)
```

Algumas transformações estão sendo realizadas no dataset.

Primeiramente, é calculada a taxa de vacinação para cada município em cada ano, dividindo a quantidade de pessoas imunizadas no ano pela população estimada do município no mesmo ano. Essa taxa é uma medida importante para avaliar a cobertura vacinal em cada município, possibilitando identificar possíveis desigualdades ou lacunas na imunização.

Em seguida, várias colunas são criadas para somar a faixa etária de homens e mulheres em diferentes grupos de idade. Essas colunas são criadas aplicando a função sum em cada linha de um conjunto de colunas que contêm as informações de população para cada faixa etária e gênero.

A próxima parte do código exclui um conjunto específico de colunas que não são relevantes para a análise. Isso é feito selecionando as colunas que começam com "HOMEM0A4" e terminam com "MULHER80", e em seguida, usando o método drop para excluir essas colunas do DataFrame.

As colunas "SOBRE60", "GINI", "IDHM" e "IDHM_L" são convertidas para o tipo float. Essas colunas contêm informações relacionadas ao Índice de Desenvolvimento Humano (IDH) e outras medidas socioeconômicas relevantes para a análise.

Por fim, algumas colunas numéricas são convertidas para o tipo "float". Essa transformação é importante para possibilitar operações matemáticas com essas variáveis, como cálculos de média, desvio padrão, correlação, entre outros.

Além disso, a substituição de vírgulas por pontos é necessária porque em alguns arquivos de dados, principalmente em arquivos gerados em sistemas operacionais com configurações regionais diferentes, a vírgula pode ser utilizada como separador decimal, enquanto em outros sistemas, o ponto é utilizado.
```python
def tratar_cobertura_vegetal(df):
    
    """
    Realizar o tratamento da coluna COBERTURA_VEGETAL_NATURAL
    Preencher os valores nulos com a média das coluns '% de cobertura vegetal natural'
    """
    
    # Selecionar apenas as colunas de interesse utilizando a função filter
    colunas_vegetal_natural = df.filter(regex='% de cobertura vegetal natural \d+')

    # Identificar as linhas com dados faltantes na coluna "COBERTURA_VEGETAL_NATURAL"
    linhas_faltantes = df['COBERTURA_VEGETAL_NATURAL'].isnull()

    # Calcular a média das colunas selecionadas para cada linha com dados faltantes
    medias = colunas_vegetal_natural.loc[linhas_faltantes].mean(axis=1)

    # Preencher os dados faltantes na coluna "COBERTURA_VEGETAL_NATURAL" com as médias calculadas
    df.loc[linhas_faltantes, 'COBERTURA_VEGETAL_NATURAL'] = medias
```

Criação da função tratar_cobertura_vegetal, que tem como objetivo preencher os valores nulos na coluna "COBERTURA_VEGETAL_NATURAL" com a média das colunas '% de cobertura vegetal natural'. Para isso, a função utiliza a função filter para selecionar apenas as colunas de interesse que contém a informação sobre a cobertura vegetal natural.

Posteriormente, a função identifica as linhas com dados faltantes na coluna "COBERTURA_VEGETAL_NATURAL" e calcula a média das colunas selecionadas para cada linha com dados faltantes.

Essa etapa é importante porque a cobertura vegetal é uma informação relevante para a análise de dados ambientais e pode influenciar na propagação da Febre Amarela de maneira indireta. Dessa forma, garantir que não haja dados faltantes nessa coluna é fundamental para uma análise mais completa e precisa dos dados.

---

```python
def tratar_concentracao_calor(df):
    
    """
    Realizar o tratamento da coluna CONCENTRACAO_FOCOS_CALOR
    Preencher os valores nulos com a média das colunas 'Concentração dos focos de calor'
    """
    
    # Selecionar apenas as colunas de interesse utilizando a função filter
    colunas_vegetal_natural = df.filter(regex='Concentração dos focos de calor \d+')

    # Identificar as linhas com dados faltantes na coluna "COBERTURA_VEGETAL_NATURAL"
    linhas_faltantes = df['CONCENTRACAO_FOCOS_CALOR'].isnull()

    # Calcular a média das colunas selecionadas para cada linha com dados faltantes
    medias = colunas_vegetal_natural.loc[linhas_faltantes].mean(axis=1)

    # Preencher os dados faltantes na coluna "COBERTURA_VEGETAL_NATURAL" com as médias calculadas
    df.loc[linhas_faltantes, 'CONCENTRACAO_FOCOS_CALOR'] = medias

tratar_cobertura_vegetal(df)
tratar_concentracao_calor(df)
```

Realizada as funções para tratar as colunas referentes à cobertura vegetal natural e concentração de focos de calor. Ambas as funções preenchem os valores nulos com a média das colunas correspondentes.

É importante realizar esse tratamento porque dados faltantes podem prejudicar a análise e interpretação dos resultados. Com os valores preenchidos com a média, é possível ter uma estimativa mais próxima da realidade e evitar distorções nos resultados finais. Além disso, a média é uma forma de agregar informações das diferentes colunas e fornecer um valor mais representativo.

### 3.5 Análise de casos por distribuição de instalações sanitárias

[Volta ao topo](#Índice-de-conteúdo)

```python
# Analise da distribuição de instalações sanitárias

categories = [
    'Rede geral de esgoto ou pluvial',
    'Fossa séptica',
    'Fossa rudimentar',
    'Vala',
    'Rio, lago ou mar',
    'Outro escoadouro',
    'Sem instalação sanitária'
]

redes = df['REDE_GERAL_ESGOTO_OU_PLUVIAL'].sum()
fossa_septica = df['FOSSA_SEPTICA'].sum()
fossa_rudimentar = df['FOSSA_RUDIMENDAR'].sum()
vala = df['VALA'].sum()
rio_lago_mar = df['RIO_LAGO_OU_MAR'].sum()
outro_escoadouro = df['OUTRO_ESCOADOURO'].sum()
sem_inst_sanitaria = df['NAO_TEM_INST_SANITARIA'].sum()

total = redes + fossa_septica + fossa_rudimentar + vala + rio_lago_mar + outro_escoadouro + sem_inst_sanitaria

percentages = [
    redes / total * 100, 
    fossa_septica / total * 100, 
    fossa_rudimentar / total * 100, 
    vala / total * 100, 
    rio_lago_mar / total * 100, 
    outro_escoadouro / total * 100, 
    sem_inst_sanitaria / total * 100
]

labels = [
    'Rede geral de esgoto ou pluvial', 
    'Fossa séptica', 
    'Fossa rudimentar', 
    'Vala', 
    'Rio, lago ou mar', 
    'Outro escoadouro', 
    'Sem instalação sanitária'
]

colors = ['#A9A9A9', '#D3D3D3', '#808080', '#696969', '#778899', '#B0C4DE', '#DCDCDC']

# Adicionando valores percentuais nas barras
for i, v in enumerate(percentages):
    ax.text(i, v+1, f'{v:.1f}%', ha='center', fontweight='bold')

# Configurações do gráfico
fig, ax = plt.subplots(figsize=(8, 4))
ax.barh(categories, percentages, color=colors)
ax.set_title(f'Distribuição de escoadouros em domicílios')
ax.set_xlabel('Percentual')
ax.set_xlim([0, 100])
ax.invert_yaxis()  # Inverter a ordem das categorias
ax.xaxis.grid(True, linestyle='--', which='major', color='grey', alpha=.25)  # Adicionar linhas de grade no eixo x
ax.yaxis.grid(False)  # Remover linhas de grade no eixo y

for i, v in enumerate(percentages):
    ax.text(v + 1, i, f'{v:.1f}%', color='black', fontweight='bold')  # Adicionar os valores das barras ao lado direito
    plt.show()
```

Em geral a grande maioria dos domicílios no Brasil possuem rede geral de esgoto com total de 85,6%.

```python
# Analise da distribuição de instalações sanitárias por Macroregião

# Agrupar por MACRORREG_LPI e criar um gráfico para cada um

region_dic = {'CO': 'CO - Região Centro-Oeste',
            'N': 'N - Região Norte',
            'S': 'S - Região Sul',
            'SE': 'SE - Região Sudeste'}

for region, data in df.groupby('MACRORREG_LPI'):
    region_name = 'Macrorregião: ' + region_dic[region]
    
    redes = data['REDE_GERAL_ESGOTO_OU_PLUVIAL'].sum()
    fossa_septica = data['FOSSA_SEPTICA'].sum()
    fossa_rudimentar = data['FOSSA_RUDIMENDAR'].sum()
    vala = data['VALA'].sum()
    rio_lago_mar = data['RIO_LAGO_OU_MAR'].sum()
    outro_escoadouro = data['OUTRO_ESCOADOURO'].sum()
    sem_inst_sanitaria = data['NAO_TEM_INST_SANITARIA'].sum()

    total = redes + fossa_septica + fossa_rudimentar + vala + rio_lago_mar + outro_escoadouro + sem_inst_sanitaria

    percentages = []
    for count in [redes, fossa_septica, fossa_rudimentar, vala, rio_lago_mar, outro_escoadouro, sem_inst_sanitaria]:
        if total == 0:
            percentage = 0
        else:
            percentage = count / total * 100
        percentages.append(percentage)
    
    # Configurações do gráfico
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.barh(categories, percentages, color=['#A9A9A9', '#D3D3D3', '#808080', '#696969', '#778899', '#B0C4DE', '#DCDCDC'])
    ax.set_title(f'Distribuição de escoadouros em domicílios ({region_name})')
    ax.set_xlabel('Percentual')
    ax.set_xlim([0, 100])
    ax.invert_yaxis()  # Inverter a ordem das categorias
    ax.xaxis.grid(True, linestyle='--', which='major', color='grey', alpha=.25)  # Adicionar linhas de grade no eixo x
    ax.yaxis.grid(False)  # Remover linhas de grade no eixo y
    for i, v in enumerate(percentages):
        ax.text(v + 1, i, f'{v:.1f}%', color='black', fontweight='bold')  # Adicionar os valores das barras ao lado direito
    plt.show()
```

Analisando os dados sobre saneamento básico nas regiões brasileiras, é possível notar que a Região Sudeste, que inclui os estados de São Paulo, Minas Gerais e Rio de Janeiro - os mais afetados pela Febre Amarela - tem o maior índice de acesso à rede geral de esgoto ou pluvial, com 86,3%. Já a Região Norte apresenta o menor índice, com apenas 1,1%.

#### Conclusão da análise de casos por distribuição de instalações sanitárias:

Embora os dados sobre saneamento básico possam ser relevantes para entender a disseminação de doenças, como a febre amarela, eles não necessariamente explicam todo os casos no dataset. Um surto de febre amarela que ocorreu entre 2017 e 2019 foi um evento complexo, que envolveu uma série de fatores, como a falta de vacinação em algumas áreas, a circulação do vírus em regiões onde não era comum.

```python
# Conversão de varíaveis para o tipo 'int64'

# Seleciona apenas as colunas do tipo 'category'
cat_cols = df.select_dtypes(include='category').columns

# Itere sobre essas colunas e converta para o tipo 'int'
for col in cat_cols:
    df[col] = df[col].cat.codes.astype('int64')
```

Neste código, eu estou convertendo as variáveis categóricas do tipo 'category' em variáveis numéricas do tipo 'int64'. Eu escolhi realizar essa conversão para poder usar essas variáveis categóricas em algoritmos de aprendizado de máquina que requerem entradas numéricas. A conversão para o tipo 'int64' garante que os valores das categorias sejam representados por números inteiros únicos e sequenciais, facilitando a análise e a modelagem dos dados.

```python
def calcula_porcentagem(df):
    colunas = ['REDE_GERAL_ESGOTO_OU_PLUVIAL', 
               'FOSSA_SEPTICA', 
               'FOSSA_RUDIMENDAR', 
               'VALA', 
               'RIO_LAGO_OU_MAR', 
               'OUTRO_ESCOADOURO', 
               'NAO_TEM_INST_SANITARIA']

    df_soma = df[colunas].sum()
    total = df_soma.sum()

    df_porcentagem = (df_soma / total) * 100
    return df_porcentagem

def classificacao(row):
    
    df_porcentagem = calcula_porcentagem(row.to_frame().T)
    rede_porcentagem = df_porcentagem['REDE_GERAL_ESGOTO_OU_PLUVIAL']
    
    return rede_porcentagem
    
df.insert(loc=18, column='CLASSIFIC_INST_SANITARIAS', value=df.apply(classificacao, axis=1))
df['CLASSIFIC_INST_SANITARIAS'] = df.apply(classificacao, axis=1)
```

Em resumo, código insere uma nova coluna no dataframe (CLASSIFIC_INST_SANITARIAS), utilizando o método insert, e aplica a função classificacao(row) em cada linha do dataframe para preencher essa coluna com a porcentagem de instalações sanitárias relacionadas à rede geral de esgoto ou pluvial.

Optei por realizar esse código para calcular e armazenar a porcentagem de instalações sanitárias relacionadas à rede geral de esgoto ou pluvial, uma vez que esse dado pode ser útil na etapa de machine learning.

```python
def plot_confusion_matrix(y_true, y_pred):
    # Definir uma paleta de cores personalizada
    cmap = ListedColormap(['lightgrey', 'salmon', 'cornflowerblue'])

    # Cria a matriz de confusão
    cm = confusion_matrix(y_true, y_pred)

    # Define o layout do gráfico
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.imshow(cm, cmap=cmap)

    # Adiciona os rótulos dos eixos
    ax.set_xlabel('Previsão')
    ax.set_ylabel('Real')
    ax.xaxis.set(ticks=(0, 1), ticklabels=('Negativo', 'Positivo'))
    ax.yaxis.set(ticks=(0, 1), ticklabels=('Negativo', 'Positivo'))

    # Adiciona os valores das células da matriz de confusão em negrito e tamanho maior
    for i in range(2):
        for j in range(2):
            ax.annotate(str(cm[i, j]), xy=(j, i), fontsize=20, fontweight='bold', color='black', ha='center', va='center')
    
    # Mostra o gráfico
    plt.show()
    print(confusion_matrix(y_true, y_pred))
    print(classification_report(y_true, y_pred))
```

```python
def cross_validation(modelo, X, y):
    # Realiza a validação cruzada
    scores = cross_val_score(modelo, X, y, cv=5)

    # Exibe a acurácia média
    print('Cross-Validation: %.2f%% (+/- %.2f%%)' % (
        scores.mean()*100, scores.std()*100))
    
# Criação de um dicionário para coletar as acurácias obtidas
acc_dict = {}
```

##### Sobre o KNN (K-Nearest Neighbors):
O KNN é um algoritmo que se baseia na distância euclidiana entre os pontos para classificar novos pontos. Ele armazena todos os pontos do conjunto de treinamento e, quando um novo ponto é apresentado ao modelo, ele calcula a distância euclidiana entre este ponto e todos os outros pontos do conjunto de treinamento. Em seguida, ele seleciona os K pontos mais próximos e faz a classificação com base na classe majoritária entre esses K pontos.

##### Vantagens:
As principais vantagens do KNN são a sua simplicidade, fácil interpretação, não parametricidade e capacidade de trabalhar com problemas de várias classes. Além disso, ele não faz suposições sobre a distribuição dos dados e pode lidar com dados não lineares, sendo uma boa opção para problemas complexos.

##### Desvantagens:
As principais desvantagens do KNN são que ele pode ser computacionalmente caro para grandes conjuntos de dados, já que ele precisa armazenar todos os pontos de treinamento. Além disso, a escolha do valor de K pode ser difícil e pode afetar significativamente o desempenho do modelo. Além disso, ele também pode ser sensível à escala dos dados, sendo necessário realizar uma normalização antes de aplicar o algoritmo.

##### Porque escolhi esse algoritmo:
Escolhi este algoritmo porque ele é fácil de entender e implementar e pode ser usado em problemas com várias classes. Além disso, ele é um algoritmo não paramétrico, ou seja, não faz suposições sobre a distribuição dos dados, o que o torna uma boa opção para problemas com dados não lineares.

```python
# Rede Neural Keras com Sequential e dropout

# Verifica o numero de colunas para imputar no modelo
num_colunas = df.shape[1] -1

# Cria o modelo
model = Sequential()
model.add(Dense(64, input_dim=num_colunas, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))

# Compila o modelo
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Treina o modelo
model.fit(X_train, y_train, epochs=150, batch_size=32, validation_data=(X_test, y_test))

# Avalia o modelo
_, accuracy = model.evaluate(X_test, y_test)
print("Acurácia do Keras: {:.2f}%".format(accuracy*100))
acc_dict['Keras'] = accuracy

# Faz a previsão com o modelo treinado
y_pred = model.predict(X_test)
y_pred = np.round(y_pred).flatten()

# Matriz de Confusão
plot_confusion_matrix(y_test, y_pred)
```

##### Sobre o Keras com Sequential e dropout:
O Keras fornece uma interface fácil de usar para criar modelos de redes neurais. O Sequential é um modelo básico do Keras que permite empilhar camadas sequencialmente para criar uma rede neural. Dropout é uma técnica de regularização que ajuda a prevenir o overfitting em modelos de redes neurais.

##### Vantagens:
- Fácil de usar: O Keras é uma biblioteca simples e fácil de usar, que facilita a criação de modelos de redes neurais.
- Flexibilidade: O Sequential do Keras permite empilhar camadas de forma sequencial, tornando-o flexível e adaptável para diferentes tipos de problemas.
- Regularização: O dropout é uma técnica de regularização que ajuda a prevenir o overfitting, melhorando o desempenho geral do modelo.
- Performance: O Keras com Sequential e dropout é conhecido por ter um desempenho rápido e eficiente em tarefas de aprendizado de máquina.

##### Desvantagens:
- Limitação na complexidade da rede: O Sequential do Keras é uma abordagem simples e básica, o que limita a complexidade da rede neural que pode ser criada.
- Pode levar ao underfitting: O uso excessivo do dropout pode levar ao underfitting, reduzindo a capacidade do modelo de aprender a partir dos dados.
- O Keras não possui uma função 'score' para ser usada na validação cruzada.

#### Por que escolhi esse algoritmo:
Eu escolhi o Keras com Sequential e dropout para o projeto de ML sobre óbitos em casos de febre amarela porque é uma abordagem simples, eficaz e fácil de usar para a criação de modelos de redes neurais. Além disso, a técnica de dropout ajuda a prevenir o overfitting, o que é importante em problemas de classificação binária como esse. A combinação dessas vantagens torna o Keras com Sequential e dropout uma escolha atraente para tarefas de aprendizado de máquina.

##### Sobre o MLP (Multilayer Perceptron):
O Multilayer Perceptron (MLP) é uma rede neural artificial (RNA) supervisionada de alimentação direta e um dos algoritmos mais populares e comumente usados em problemas de classificação. A estrutura do MLP é composta por uma ou mais camadas ocultas de neurônios, além de uma camada de entrada e uma camada de saída.

##### Vantagens:
O MLP é capaz de lidar com problemas não lineares complexos, e sua estrutura em camadas ocultas permite a extração de recursos úteis para a classificação dos dados.
É capaz de trabalhar com grandes quantidades de dados e pode lidar com problemas com múltiplas classes.
O MLP é altamente flexível e pode ser personalizado para diferentes tarefas de classificação por meio da escolha da função de ativação, número de camadas e neurônios em cada camada, e técnicas de regularização.

##### Desvantagens:
O treinamento do MLP pode ser demorado e requer grandes quantidades de dados para obter resultados precisos. Além disso, a escolha inadequada dos parâmetros do MLP pode levar a overfitting ou underfitting dos dados.
O MLP é altamente dependente da qualidade dos dados de entrada e pode ser afetado por dados ausentes, ruidosos ou mal rotulados. Além disso, o MLP pode ser sensível a outliers e pode levar a resultados imprecisos.
A interpretação dos resultados do MLP pode ser difícil, pois a rede neural não fornece insights sobre como os dados estão sendo classificados.

##### Porque escolhi esse algoritmo:
